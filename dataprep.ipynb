```python

from sqlalchemy import create_engine, text
import logging

# üîß Set up basic logging
logging.basicConfig(level=logging.INFO)
logger = logging.getLogger(__name__)

try:

    
    # üîå Connect to PostgreSQL
    DATABASE_URL = "postgresql://postgres:root@localhost:5432/food_waste_manage"
    engine = create_engine(DATABASE_URL,echo=True)

    # üìú Define SQL queries
    # the table column will insert with lowercase.
    queries = [
        """CREATE TABLE IF NOT EXISTS providers_data (
            Provider_ID SERIAL PRIMARY KEY,
            Name VARCHAR(100),
            Type VARCHAR(100),
            Address VARCHAR(100),
            City VARCHAR(100),
            Contact VARCHAR(100)
        );""",
        
        """CREATE TABLE IF NOT EXISTS receivers_data (
            Receiver_ID SERIAL PRIMARY KEY,
            Name VARCHAR(100),
            Type VARCHAR(100),
            city VARCHAR(100),
            Contact VARCHAR(100)
        );""",
        
        """CREATE TABLE IF NOT EXISTS food_listings_data (
            Food_ID SERIAL PRIMARY KEY,
            Food_Name VARCHAR(100),
            Quantity INT,
            Expiry_Date DATE,
            Provider_ID INT REFERENCES providers_data(Provider_ID),
            Provider_type TEXT,
            Location VARCHAR(100),
            Contact VARCHAR(100)
            meal_type VARCHAR(100)
        );""",
        
        """CREATE TABLE IF NOT EXISTS claims_data (
            Claim_ID SERIAL PRIMARY KEY,
            Food_ID INT REFERENCES food_listings_data(Food_ID),
            Receiver_ID INT REFERENCES receivers_data(Receiver_ID),
            Status VARCHAR(30),
            Timestamp TIMESTAMP WITHOUT TIME ZONE
        );"""
    ]
    query3= """CREATE TABLE IF NOT EXISTS claims_data (
            Claim_ID SERIAL PRIMARY KEY,
            Food_ID INT REFERENCES food_listings_data(Food_ID),
            Receiver_ID INT REFERENCES receivers_data(Receiver_ID),
            Status VARCHAR(30),
            Timestamp TIMESTAMP WITHOUT TIME ZONE
        );"""
    
    # üß† Execute queries safely
    with engine.connect() as conn:
       # for query in queries:
           conn.execute(text(query3))
           conn.commit()
           logger.info("Executed query successfully.")
    
    print("‚úÖ Tables created in PostgreSQL!")

except FileNotFoundError as e:
    logger.error(f"üìÅ File not found: {e}")
except Exception as e:
    logger.error(f"üí• An error occurred: {e}")
```
